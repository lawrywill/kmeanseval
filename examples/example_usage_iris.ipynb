## Setup

```python
import kmeanseval
import pandas as pd
from sklearn import datasets
```

```python
# load dataset
df = datasets.load_iris()
df = pd.DataFrame(df['data'])

# define range for k
rng = range(2, 11)
```

## Calculate metrics

```python
# instantiate the KMeansEvaluator object
kme = kmeanseval.KMeansEvaluator(data = df, k_range = rng, init = 'k-means++')
```

```python
# check that models are created
kme._models
```

```python
# get within-cluster-sum of squared errors
wss = kme.get_metrics(metric = 'wss')
wss
```

```python
# get average silhouette scores
sscore = kme.get_metrics(metric = 'silhouette_score')
sscore
```

```python
# get sample silhouette coefficients
ssample = kme.get_metrics(metric = 'silhouette_sample')
ssample
```

```python
# test default behavior
default = kme.get_metrics(metric = 'oops')
default
```

## Plot metric values

```python
# plot the WSS elbow
kme.plot_elbow(figsize = (15, 8), textsize = 15)
```

```python
# plot the average silhouette score
kme.plot_avg_silhouette_scores(figsize = (15, 8), textsize = 15)
```

```python
# plot the full set of sample silhouette scores compared to the average
kme.plot_silhouette_scores(figsize = (15, 8), textsize = 15)
```
